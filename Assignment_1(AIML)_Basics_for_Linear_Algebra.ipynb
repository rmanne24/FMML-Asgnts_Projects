**The multi_dot method**

The multi_dot function is a function that is a present in the NumPy Linear Algebra module.

It helps in performing matrix multiplication between more than two matrices.

Here, we are storing arrays in the variables. np.random.randint function is responsible for generating random numbers.

-4,4 here is used to specify the range of number we are interested in generating. Here we are expecting that the function would print numbers between -4 to 3.

And here 500,5 is the representation of the shape of the matrix

```python
 # matrices with random integers: entries ranging from -4 to 4
 a = np.random.randint(-4,4,(500,5))
 b = np.random.randint(-4,4,(5,1000))
 c = np.random.randint(-4,4,(1000,10))
 d = np.random.randint(-4,4,(10,2000))
 e = np.random.randint(-4,4,(2000,200))
 # Perform multiple matrix multiplication
 z = np.linalg.multi_dot([a,b,c,d,e])
 print(z.shape)
 print(z)
```

here np.linalg.multi_dot([a,b,c,d,e]) is equivalent to z=(((((a⋅b)⋅c)⋅d)⋅e))

here we are printing the shape i.e. its dimesnions and the matrix z.

The result of this method can be obtained with successive dot products of matrices but multi_dot functions in an optimized manner. It decides the order of dot multiplication to complete the entire process efficiently.

```python
%%time
z = np.linalg.multi_dot([a,b,c,d,e])
print(z, '\n')
```

%%time is a function used to calculate the amount of time taken to run a particular cell.

```python
 %%time
 z = a.dot(b).dot(c).dot(d).dot(e)
 print(z, '\n')
```

when cpu time is compared for dot method and multi_dot method we see that multi_dot is more efficient

**Outer Product**

In outer product each element of a is multiplied with each element of b.

Here in the given example we would be mutilplying 5X1 matrix with 3X1 matrix we get 5X3 matrix since we would be performing an outer product on it.


More Generalized statement:
Outer product is the dot product of a column vector of size Mx1 and a row vector of size 1xN. The resulting array is a matrix of size MxN.

```python
 a = np.array([1,2,3,4,5])
 b = np.array([6,3,2])
 z = np.outer(a,b)
 print(z)
```

**Solving System of Equations**

We can solve the system of linear equations using numpy

Here we are storing the coefficient matrix in a and variable matrix in x and the dependent varible in b and we use x = np.linalg.solve(a,b) and get the solutions of the bunch of equations.

```python
# Coefficient Matrix
a = np.random.randint(1,20,[4,4])
# Dependent variable vector
b = np.array([4,9,12,7])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
# Check for correctness
B = a.dot(x)
print(B)

# This ‘B’ array is identical to the input ‘b’ array. Hence, our solution is correct.
```

here we are just performing dot product multiplication between a and x which gives b in return, we are verifying it.

**Singular Value Decomposition**

SVD or singular value decomposition is used for various purposes in AI & ML.

SVD states that a matrix A can be represented as A = U.S.Vt



skimage is a package in python that is used to image processing.

SVD is used to compress the data by removing the less important features

Dimensionality reduction

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
image = data.astronaut()
print(image.dtype, image.min(), image.max(), image.shape)
# Show original image
plt.imshow(image)
plt.title("Original Image")
plt.axis('on')
plt.show()
```

Using the above code we were able to display an image of an astronaut from skimage library, an rgb image.

```python
# Normalize pixel values to [0, 1]
img = image / 255.0

# Reorder axes: channels first (C, H, W)
img = np.transpose(img, axes=(2, 0, 1))
print(img.shape)  # (3, 512, 512)
```

To perform SVD on image, an image can be broken down into 3 different components U matrix, Svector and V matrix.

The S vector is just the diagonal entries of the sigma matrix.

We must know that the first dimension 3 refers to the 3 colour channels.

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

The 3 colour channels are printed using this program.

```python
# S matrix should have dimensions suitable for matrix multiplication
Sigma = np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

**Dimensionality reduction**

Here, we reconstructed the image with most details, even at a reduction of ¼.

We can once again reconstruct the same image by reducing the data points from 512 to 20.

```python
k = 20
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
